```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Passo 1: Importação e Carregamento dos Dados
df = pd.read_csv('/home/coldrand/pyTBI-data/data/Rates_of_TBI-related_Emergency_Department_Visits__Hospitalizations__and_Deaths_____United_States__2001___2010.csv')

```

```python
# Passo 2: Pré-processamento dos Dados
# Verificar valores ausentes
print(df.isnull().sum())

# Preencher valores ausentes ou remover registros incompletos
df = df.dropna()
```

```python
# Passo 3: Análise Exploratória de Dados (EDA)
# Resumo estatístico dos dados
print(df.describe())

#Hospitalizações vs ano
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Hospitalizations'], marker='o')

# Calcular a linha de tendência
z = np.polyfit(df['Year'], df['Hospitalizations'], 3)
p = np.poly1d(z)
plt.plot(df['Year'], p(df['Year']), linestyle='--', color='red', label='Linha de Tendência')

# Títulos e rótulos
plt.title('Tendência das Hospitalizações ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Hospitalizações')
plt.grid(True)
plt.legend()
plt.show()

```

```python
# Passo 4: Visualização de Dados
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Hospitalizations', y='Deaths', hue='Year', palette='viridis')
plt.title('Dispersão entre Hospitalizações e Mortes')
plt.xlabel('Hospitalizações')
plt.ylabel('Mortes')
plt.show()
```
